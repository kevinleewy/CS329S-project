```python
import json
import matplotlib.pyplot as plt
import numpy as np
import os
import pandas as pd
import random

p = inflect.engine()
```

```python
base_path = '../data/In-shop Clothes Retrieval Benchmark'
splits_file = os.path.join(base_path, 'Eval/list_eval_partition.txt')
```

```python
splits = pd.read_csv(splits_file, sep='\s+', header=1)
splits['sex'] = splits.apply(lambda row: row['image_name'].split('/')[1], axis=1)
splits['category'] = splits.apply(lambda row: row['image_name'].split('/')[2], axis=1)
splits
```

```python
gallery_images = splits[splits['evaluation_status']=='gallery']
gallery_images
```

```python
gallery_items = gallery_images.groupby('item_id')
len(gallery_items.groups)
```

```python
gallery_front_images = gallery_images[gallery_images['image_name'].str.contains('front')]
print(f'{len(gallery_front_images)} out of {len(gallery_images)} images are front images')
print(f'{len(gallery_front_images)} out of {len(gallery_items.groups)} items have front images')
gallery_front_images
```

```python
gallery_front_images_list = gallery_front_images.values.tolist()
gallery_front_images_list
```

```python
def toCatalogObject(x):
    
    sex = x[3][0] + x[3][1:].lower()
    category = x[4].replace('_', ' ')
    return {
        "Label": f'{sex}\'s {p.singular_noun(category)}',
        "Category": category,
        "Sex": sex,
        "Rating": random.randint(3,5),
        "Price": "${:.2f}".format(random.gauss(50, 20)),
        "URI": x[0]
    }
catalog = list(map(toCatalogObject, gallery_front_images_list))
```

```python
catalog
```

```python
with open('catalog.json', 'w') as outfile:
    json.dump(catalog, outfile, indent=4)
```
